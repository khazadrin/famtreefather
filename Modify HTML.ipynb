## Csv to df \ name of main file \ git proj link

```python
NAME = 'Leonid_Alipiev'
GITHTMLSOURCE = 'https://khazadrin.github.io/famtreefather/'
```

```python
import pandas as pd
from os import getcwd
from bs4 import BeautifulSoup

__file__ = getcwd()

df = pd.read_csv(f'{__file__}/{NAME}.csv', index_col=0)

df.head(3)
```

## Main html modify

```python
html = open(f'{__file__}/{NAME}.html', encoding='utf-8')
soup = BeautifulSoup(html, 'html.parser')
#Leonid_Alipiev.html.files/p' + i + '.html?s=40e61edcbe0f6833
```

### Link of photoes to medium size photoes

```python
#<image id="i261" 
#       xlink:href="Leonid_Alipiev.html.files/io764_13_920.jpg" 
#       x="1" 
#       y="1" 
#       width="60" 
#       height="13.8000001907349" 
#       preserveAspectRatio="none"/>
for tag in soup.find_all('image'):
    if tag.has_attr('xlink:href'):
        tag['xlink:href'] = tag['xlink:href'].removeprefix(f'{NAME}.html.files/')
        print(tag['xlink:href'])
        if not df[df['Filename'] == tag['xlink:href']].empty:
            tag['xlink:href'] = df[df['Filename'] == tag['xlink:href']]['M'].iloc[0]
```

### Link of htmls ro git hub htmls

```python
#<g at:id="956" 
#      class="s5" 
#      id="g28" 
#      onclick="nodeclick(956)">
for tag in soup.find_all('g'):
    if tag.has_attr('onclick'):
        print(tag['onclick'])
        full_tag = tag['onclick'].removeprefix('nodeclick(').removesuffix(')')
        if (full_tag.isdigit()):
            full_tag = 'p'+ full_tag + ".html"
            tag['onclick'] = f"nodeclick('{GITHTMLSOURCE}{full_tag}')"
```

### Save changes

```python
html = soup.prettify("utf-8")
with open(f'{__file__}/{NAME}.html', "wb") as file:
    file.write(html)
```

## Modify htmls connected to Main html

```python
from os import listdir
from os.path import isfile, join
__htmlfile__ = f"{__file__}\{NAME}.html.files"
htmlfiles = [f for f in listdir(__htmlfile__) if isfile(join(__htmlfile__, f)) and (f.find('.html'))>-1]
htmlfiles[0]
```

```python
for file in htmlfiles:
    html = open(f"{__htmlfile__}\{file}", encoding='utf-8')
    soup = BeautifulSoup(html, 'html.parser')
    for tag in soup.find_all("a"):
        if (tag.has_attr("href")):
            if not (df[df['Filename'] == tag['href']]).empty:
                tag['href'] = df[df['Filename'] == tag['href']]['File Link'].iloc[0]
            for item in [f for f in df['Filename'] if f.find('.html') >= 0]:
                if tag['href'].find(item) >= 0:
                    tag['href'] = f"{GITHTMLSOURCE}{item}"
```

```python
#<a href="io4.jpg" 
#   target="_blank">
#<img width="100%" 
#   src="io4.jpg" 
#   alt="Леонид Алипиевич" /></a>

for file in htmlfiles:
    html = open(f"{__htmlfile__}\{file}", encoding='utf-8')
    soup = BeautifulSoup(html, 'html.parser')
    for tag in soup.find_all('a'):
        if (tag.has_attr("href")):
            if not (df[df['Filename'] == tag['href']]).empty:
                tag['href'] = df[df['Filename'] == tag['href']]['File Link'].iloc[0]
            for item in [f for f in df['Filename'] if f.find('.html') >= 0]:
                if tag['href'].find(item) >= 0:
                    tag['href'] = f"{GITHTMLSOURCE}{item}"
    for tag in soup.find_all('img'):
        if (tag.has_attr("src")):
            if not (df[df['Filename'] == tag['src']]).empty:
                tag['src'] = df[df['Filename'] == tag['src']]['File Link'].iloc[0]
    soup = soup.prettify("utf-8")
    with open(f'{__htmlfile__}/{file}', "wb") as html:
        html.write(soup)
```
